## Prepare Your Dataframe

Import any libraries you need and create a dataframe.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

my_df = pd.read_csv('/Users/vasanthibodasingi/Desktop/LaunchCode/Womens+/netflix_titles_cleaned.csv')

print(my_df.info())
```

## Find Correlations

Use either pairplot or a heatmap or both to find the two variables with the strongest correlation in your dataset.

show_id and release_year

```python
# Calculating the correlation matrix
corr_matrix = my_df.corr()

# Creating heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

## Create Your Model

Use the two columns with the strongest correlation to create and train your model. Make sure to print out the summary and plot the column values and the line produced by the model.

```python
# Identifying the two columns with the strongest correlation
strongest_correlation_columns = my_df.corr().unstack().sort_values(ascending=False)
strongest_correlation_columns = strongest_correlation_columns[strongest_correlation_columns != 1.0]
strongest_corr_column1, strongest_corr_column2 = strongest_correlation_columns.index[0]

# Extract the data for the two columns
X = my_df[[strongest_corr_column1]]
y = my_df[strongest_corr_column2]

# Create and train the linear regression model
model = LinearRegression()
model.fit(X, y)

# model summary
print("Model Coefficients:")
print("Intercept (b0):", model.intercept_)
print(f"Coefficient (b1) for {strongest_corr_column1}: {model.coef_[0]}")

# Plot the data and the line produced by the model
plt.scatter(X, y, label="Data Points")
plt.plot(X, model.predict(X), color='red', label="Regression Line")
plt.title(f"Linear Regression: {strongest_corr_column2} vs. {strongest_corr_column1}")
plt.xlabel(strongest_corr_column1)
plt.ylabel(strongest_corr_column2)
plt.legend()
plt.show()
```

## Error Terms

Finally, plot your error terms!

```python
# Calculating error terms
residuals = y - model.predict(X)

# Ploting error terms
plt.figure(figsize=(8, 6))
plt.scatter(X, residuals, color='blue', alpha=0.6)
plt.axhline(y=0, color='red', linestyle='--', linewidth=2, label="Residuals Mean")
plt.title("Residual Plot")
plt.xlabel(strongest_corr_column1)
plt.ylabel("Residuals")
plt.legend()
plt.show()

```
